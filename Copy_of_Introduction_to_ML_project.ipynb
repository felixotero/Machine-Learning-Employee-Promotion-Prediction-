```python
#preprocessing
# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
employee_set.department=le.fit_transform(employee_set.department)
employee_set.region=le.fit_transform(employee_set.region)
employee_set.education=le.fit_transform(employee_set.education)
employee_set.gender=le.fit_transform(employee_set.gender)
employee_set.recruitment_channel=le.fit_transform(employee_set.recruitment_channel)

#dropping columns
employee_set = employee_set.drop(["employee_id"], axis=1)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

employee_df_train, employee_df_valid = train_test_split(employee_set, test_size=0.25, random_state=12345)

train_features = employee_df_train.drop(["is_promoted"], axis=1)
train_target = employee_df_train["is_promoted"]

test_features = employee_df_valid.drop(["is_promoted"], axis=1)
test_target = employee_df_valid["is_promoted"]

```

```python
train_features['department'].describe()
```

```python
model = DecisionTreeClassifier(random_state=111)
model.fit(train_features,train_target)
model.predict(test_features)
```
